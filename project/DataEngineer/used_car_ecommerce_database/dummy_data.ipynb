## Table city, province and users

```python
# Create function to return city latitude and longitude
def mycity_locator(city):
    geolocator = Nominatim(user_agent='city_locator')
    city_coordinates = []
    try:
        location = geolocator.geocode(city)
        if location:
            city_coordinates = [location.latitude, location.longitude]
        else:
            city_coordinates = ["Not found", "Not found"]
        time.sleep(1)
    except Exception as e:
        city_coordinates = [f"Error: {e}", f"Error: {e}"]
    return city_coordinates
```

```python
set_seed()

city = pd.DataFrame()
province = pd.DataFrame()
users = pd.DataFrame()

n_city = 60
n_province = 20
n_users = 200

city_key = [60001+i for i in range(n_city)]
province_key = [70001+i for i in range(n_province)]
users_key = [80001+i for i in range(n_users)]

# table city
city['city_id'] = city_key
city['city'] = [fake_id.unique.city() for i in range(n_city)]
city['province_id'] = [np.random.choice(np.array(province_key)) for i in range(n_city)]
city['coordinate'] = city['city'].apply(lambda x: tuple(mycity_locator(x)))
city['date_created'] = [fake_id.date_between(start_date='-6y', end_date='-5y') for i in range(n_city)]

# table province
province['province_id'] = province_key
province['province'] = [fake_id.unique.state() for i in range(n_province)]
province['country_id'] = [10001 for i in range(n_province)]
province['date_created'] = [fake_id.date_between(start_date='-6y', end_date='-5y') for i in range(n_province)]

# table users
users['user_id'] = users_key
users['user_name'] = [fake_id.unique.name() for i in range(n_users)]
users['password'] = [fake_id.password(length=8, special_chars=True,digits=True, upper_case=True,lower_case=True) for i in range(n_users)]
users['phone_number'] = [fake_id.phone_number() for i in range(n_users)]
users['social_media_address'] = [''.join(['instagram.com/', i.lower().replace('.','').replace(' ','_').replace(',','')]) for i in users['user_name']]
users['domicile_address'] = [fake_id.address() for i in range(n_users)]
users['domicile_address'] = users['domicile_address'].str.replace('\n',',')
users['city_id'] = [np.random.choice(np.array(city_key)) for i in range(n_users)]
users['date_created'] = [fake_id.date_between(start_date='-5y', end_date='now') for i in range(n_users)]

display(city)
display(province)
display(users)
```

## Table ads

```python
set_seed()

ads = pd.DataFrame()
n_ads = 70

ads_key = [90001+i for i in range(n_ads)]

ads['ads_id'] = ads_key
ads['title'] = [fake_id.text()[:10] for i in range(n_ads)]
ads['description'] = [fake_id.text() if len(fake_id.text()) <= 30 else fake_id.text()[:30] for i in range(n_ads)]
ads['user_id'] = [np.random.choice(np.array(users_key)) for i in range(n_ads)]
ads = ads.merge(users[['user_id', 'date_created']], how='left', on='user_id')
ads['date_created'] = pd.to_datetime(ads['date_created'])
ads['date_created'] = [i + pd.Timedelta(days=np.random.randint(1,32)) for i in ads['date_created']]
ads['date_created'] = ads['date_created'].dt.strftime('%Y-%m-%d')

display(ads)
```

## Table bid, car and model

```python
set_seed()

model = pd.DataFrame()
car = pd.DataFrame()
bid = pd.DataFrame()

n_model = 40
n_car = 200
n_bid = 1000

model_key = [11001+i for i in range(n_model)]
car_key = [12001+i for i in range(n_car)]

model['model_id'] = model_key
model['model_name'] = model_list
model['transmission_id'] = [np.random.choice(np.array(transmission_key)) for i in range(n_model)]
model['body_id'] = [np.random.choice(np.array(body_key)) for i in range(n_model)]
model['brand'] = [i.split(' ')[0] for i in model_list]
model = model.merge(brand[['brand_id', 'brand']], how='left', on='brand')
model.drop(columns='brand', inplace=True)

car['car_id'] = car_key
car['permit_bid'] = [fake_id.boolean(chance_of_getting_true = 50) for i in range(n_car)]
car['color'] = [fake_id.color_name() for i in range(n_car)]
car['mileage_km'] = [np.random.randint(10,100)*10000 for i in range(n_car)]
car['mfg_year_id'] = [np.random.choice(np.array(mfg_year_key)) for i in range(n_car)]
car['model_id'] = [np.random.choice(np.array(model_key)) for i in range(n_car)]
car['ads_id'] = [ads_key[i%len(ads_key)] for i in range(n_car)]
car['price'] = [np.random.randint(10,200)*10000000 for i in range(n_car)]
car = car.merge(ads[['ads_id', 'date_created']], how='left', on='ads_id')

model = model.merge(car[['model_id', 'date_created']], how='left', on='model_id')
model['date_created'] = model['date_created'].fillna('2018-12-05')
model['date_created'] = pd.to_datetime(model['date_created'])
model = model.sort_values(by='date_created', ascending=True)
model = model.drop_duplicates(subset='model_id', keep='last')
model = model.reset_index(drop=True)

car_bid_key = list(car[car['permit_bid']==True]['car_id'].unique())
bid['bid_id'] = [13001+i for i in range(n_bid)]
bid['bid_price'] = [np.random.randint(10,200)*10000000 for i in range(n_bid)]
bid['user_id'] = [np.random.choice(np.array(users_key)) for i in range(n_bid)]
bid['car_id'] = [np.random.choice(np.array(car_bid_key)) for i in range(n_bid)]
bid = bid.merge(car[['car_id', 'date_created']], how='left', on='car_id')
bid['date_created'] = pd.to_datetime(bid['date_created'])
bid['date_created'] = [i + pd.Timedelta(days=np.random.randint(1,101)) for i in bid['date_created']]
bid['date_created'] = bid['date_created'].dt.strftime('%Y-%m-%d')

display(car)
display(model)
display(bid)
```
